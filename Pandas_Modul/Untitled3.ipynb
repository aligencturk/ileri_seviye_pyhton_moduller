```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
 # DataFrame oluşturma (kodlar)

stn_boy = np.random.randint(150,190,(10))
stn_kilo = np.random.randint(50,100,(10))
stn_no = np.random.randint(2100,2200,(10))

# pandas kütüphanesi , dictionary veri yapısı üzerine kurulmuştur 
# pandas kütüphanesi ile excel ve sql deki her şeyi yapabiliriz

my_dict = {"ogr_no":stn_no,"ogr_boy":stn_boy,"ogr_kilo":stn_kilo}

my_df = pd.DataFrame.from_dict(my_dict)
my_df
```

```python
my_df.index
```

```python
my_df.index = ["ogrenci_1","ogrenci_2","ogrenci_3","ogrenci_4","ogrenci_5","ogrenci_6","ogrenci_7","ogrenci_8","ogrenci_9","ogrenci_10"]
my_df
```

```python
# dataframe ye yeni sütun ve değerler ekleme

stn_vize = np.random.randint(20,100,(10))
stn_final = np.random.randint(20,100,(10))

my_df["ogr_vize"] = stn_vize
my_df["ogr_final"] = stn_final
my_df
```

```python
deger_vize = my_df["ogr_vize"].values
deger_final = my_df["ogr_final"].values
```

```python
# ogr_gecti isimli bir key ekleyn
# bu keyin değerleri vize notlarının yüzde kırkı ile
#final notlarının yüzde altmışının toplamı olucak

son_not= (my_df["ogr_vize"].values*0.4) + (my_df["ogr_final"].values*0.6)
my_df ["ogr_gecti"] = son_not




my_df



```

```python
my_df["durum"] = my_df["ogr_gecti"].values>50
my_df
```

```python
my_df["cinsiyet"]="erkek"  #şart koyulmadığı için tüm sütunun adını erkek yaptı
my_df
```

```python
#istenilen özelliklere göre değer getirtme
# tek sorguya göre
my_df[my_df["ogr_vize"]<50]
```

```python
#birden fazla sorguya göre değer çağırma
my_df[(my_df["ogr_vize"]<50) & (my_df["durum"]==True)]
```
